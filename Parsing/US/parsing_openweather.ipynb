```python
import json
from datetime import datetime
import pandas as pd
import os
```

```python
def split_date(inp):
    return inp.split('.json')[0].split('_')[0]
```

```python
def split_date_date(inp):
    date_string = inp.split('.xml')[0].split('_')[1]
    
    month = int(date_string[0:2])
    day = int(date_string[2:4])
    year = int(date_string[4:8])
    
    return datetime(year,month,day)
```

```python
columns = ['Location','Date','Date of Prediction','Precipitation']
```

```python
df = pd.DataFrame(columns = columns)
```

```python
path = "C:/Users/u/Dropbox/Final Year Project/Python/us_openweather"

for filename in os.listdir(path):
    if '.json' in filename:
        with open(filename,'r') as json_file:
            print(filename)
            data = json.load(json_file)
            
            for i in range(0,len(data['hourly'])):
                try:
                    #print(datetime.utcfromtimestamp(data['hourly'][i]['dt']).strftime('%Y-%m-%d %H:%M:%S'))
                    #print(data['hourly'][i]['rain']['1h'])
                    df = df.append({'Location':split_date(filename),
                                   'Date':datetime.utcfromtimestamp(data['hourly'][i]['dt']).strftime('%Y-%m-%d %H:%M:%S'),
                                    'Date of Prediction':split_date_date(filename),
                                   'Precipitation':data['hourly'][i]['rain']['1h']},ignore_index=True)
                except:
                    #print('0')
                    df = df.append({'Location':split_date(filename),
                                    'Precipitation':0.0,
                                    'Date of Prediction':split_date_date(filename),
                                'Date':datetime.utcfromtimestamp(data['hourly'][i]['dt']).strftime('%Y-%m-%d %H:%M:%S')},ignore_index=True)
```
