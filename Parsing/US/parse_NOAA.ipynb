```python
import xmltodict
import json
import os
import pandas as pd
import datetime
from datetime import timedelta
import qgrid

qgrid.enable()
```

```python
def split_date(inp):
    
    date_string = inp.split('.xml')[0].split('_')[1]
    
    month = int(date_string[0:2])
    day = int(date_string[2:4])
    year = int(date_string[4:8])
    
    return datetime.datetime(year,month,day)
```

```python
def split_name(inp):
    return inp.split('.xml')[0].split('_')[0]
```

```python
columns = ['Location','Date of Prediction','Date','Precipitation','Probability']
```

```python
df = pd.DataFrame(columns = columns)
```

```python
hourly_df = pd.DataFrame(columns = columns)
```

```python
path = "C:/Users/u/Dropbox/Final Year Project/Python/NOAA"

for filename in os.listdir(path):
    if '.xml' in filename:
        try:
            with open(filename,'r') as json_file:
                    print(filename)
                    data = json.load(json_file)
        
                    for i in range(0,len(data['properties']['quantitativePrecipitation']['values'])):
                        df = df.append({'Location':split_name(filename),
                        'Date of Prediction': split_date(filename),
                        'Date':str(data['properties']['quantitativePrecipitation']['values'][i]['validTime']),
                        'Precipitation':data['properties']['quantitativePrecipitation']['values'][i]['value'],
                        'Probability':data['properties']['probabilityOfPrecipitation']['values'][i]['value']},ignore_index=True)
        except Exception as e:
            print(e)
```

```python
# For each row check the number N after the timestamp. If N equals 1, skip
# Append N new rows with values/N
# Delete the original row
# This results in a dataframe which has averaged the aggregated values across the time grain

for i in range(0,len(df.index)):
    try:
        aggregation_level = int(df['Date'].iloc[i].split('/')[1][2])
        #print(aggregation_level)

        # Create a row with the value for each times it has been aggregated
        for j in range(0,aggregation_level):
            hourly_df = hourly_df.append({
                'Location':df['Location'].iloc[i],
                'Date of Prediction':df['Date of Prediction'].iloc[i],
                'Date':str(pd.to_datetime(df['Date'].iloc[i].split('/')[0]) + timedelta(hours = j)),
                'Precipitation':float(df['Precipitation'].iloc[i])/float(aggregation_level),
                'Probability':df['Probability'].iloc[i]},ignore_index=True
                                )
            
            
    except Exception as e:
        print(e)
        #df = df.append('Date':df['Date'].iloc[i],
                      #'Time':)
```

```python
hourly_df.to_csv("NOAA_forecast.csv",index=False)
```
