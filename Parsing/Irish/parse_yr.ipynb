```python
import json
import pandas as pd
import os
import datetime
```

```python
columns = ['Location','Number','Date of Prediction','Date','Weather Description','Precipitation']
```

```python
def split_location(inp):
    return inp.split('.json')[0].split('_')[0]
```

```python
def split_date(inp):
    
    date_string = inp.split('.xml')[0].split('_')[1]
    
    month = int(date_string[0:2])
    day = int(date_string[2:4])
    year = int(date_string[4:8])
    
    return datetime.datetime(year,month,day)
```

```python
df = pd.DataFrame(columns = columns)
```

```python
path = "C:/Users/u/Dropbox/Final Year Project/Python/yr"

for filename in os.listdir(path):
    if '.json' in filename:
        try:
            with open(filename,'r') as json_file:
                print(filename)
                data = json.load(json_file)
                try:
                    for i in range(0,len(data['properties']['timeseries'])):
                        df = df.append({'Location': split_location(filename),
                        'Number':i ,
                        'Date of Prediction':split_date(filename),
                        'Date':data['properties']['timeseries'][i]['time'],
                        'Weather Description':data['properties']['timeseries'][i]['data']['next_1_hours']['summary']['symbol_code'],
                        'Precipitation':data['properties']['timeseries'][i]['data']['next_1_hours']['details']['precipitation_amount']},ignore_index=True)
                except:
                    print("No more Data on this day. Got to " + str(i))
        except:
            print("Error")
```
