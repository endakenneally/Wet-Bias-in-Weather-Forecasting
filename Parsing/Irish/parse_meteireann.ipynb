```python
def split_date(inp):
    
    date_string = inp.split('.xml')[0].split('_')[1]
    
    month = int(date_string[0:2])
    day = int(date_string[2:4])
    year = int(date_string[4:8])
    
    return datetime.datetime(year,month,day)
```

```python
def return_location(inp):
    return inp.split('.xml')[0].split('_')[0]
```

```python
columns = ['Location','Date of Prediction','Date','Finish','Precipitation'] 
df = pd.DataFrame(columns = columns)
```

```python
start_time = time.time()

path = "C:/Users/u/Dropbox/Final Year Project/Python/met_eireann"

for filename in os.listdir(path):
    if '.xml' in filename:
#         print(filename)
        try:
        #First we open the xml file and dump to a dictionary
            with open(filename) as xml_file:
                #print(filename)
                data_dict = xmltodict.parse(xml_file.read())
                xml_file.close()
                json_data = json.dumps(data_dict)
                data = json.loads(json_data)
                
            try:
                for i in range(1,len(data['weatherdata']['product']['time']),2):
                    df = df.append({'Location':return_location(filename),
                    'Date of Prediction': split_date(filename),
                    'Date': data['weatherdata']['product']['time'][i]['@from'],
                    'Finish':data['weatherdata']['product']['time'][i]['@to'],
                    'Precipitation':data['weatherdata']['product']['time'][i]['location']['precipitation']['@value']},ignore_index=True)
            except:
                print(filename+" : error")
        except:
            print(filename+" : xml error")
            
print("--- %s seconds ---" % (time.time() - start_time))
```
